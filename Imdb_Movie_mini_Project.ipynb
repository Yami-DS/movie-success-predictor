```python
# What Makes a Movie Successful?

This mini data analysis project explores how IMDb ratings and movie duration relate to box office revenue. Using a small dataset and simple visuals, I aimed to answer:
**Do better-rated or longer movies make more money?**
```

## Importing required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
df = pd.read_csv("movies.csv")
df
```

## Basic info

```python
df.info()
```

```python
df.describe()
```

## Scatter Plot- Imdb Rating vs Gross

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x="IMDB_Rating",y="Gross_Million_USD",hue="Movie_Name")
plt.title("IMDb Rating vs Box Office Gross")
plt.xlabel("IMDb Rating")
plt.ylabel("Gross (in Million USD)")
plt.grid(True)
plt.show()
```

## Scatter Plot - Duration vs Gross

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Duration_Minutes", y="Gross_Million_USD", hue="Movie_Name")
plt.title("Movie Duration vs Box Office Gross")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Gross (in Million USD)")
plt.grid(True)
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Key Insights

```python
print(" Insights:")
print("- Movies with higher IMDb ratings generally earn more revenue.")
print("- Duration has a mild positive effect on gross revenue.")
print("- Some exceptions exist (e.g., Titanic's huge gross despite moderate rating).")
```
